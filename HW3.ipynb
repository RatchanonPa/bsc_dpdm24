```python
# Define top 10% as High-Value Customers
high_value_threshold = customer_total_spending.quantile(0.9)
high_value_customers = customer_total_spending[customer_total_spending >= high_value_threshold].index
```

```python
# Define holiday period
holiday_data = data_test[
    ((data_test['month'] == 12) & (data_test['day'].between(15, 31))) |
    ((data_test['month'] == 1) & (data_test['day'].between(1, 10)))
]

# Separate High-Value and Regular Customers during holiday period
high_value_holiday_purchases = holiday_data[holiday_data['CustomerID'].isin(high_value_customers)]
regular_holiday_purchases = holiday_data[~holiday_data['CustomerID'].isin(high_value_customers)]
```

```python
import mlxtend
print(mlxtend.__version__)
```

```python
# Use Apriori algorithm for association rules
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

# Prepare data for high-value customers
high_value_transactions = high_value_holiday_purchases.groupby('InvoiceNo')['Description'].apply(list)
regular_transactions = regular_holiday_purchases.groupby('InvoiceNo')['Description'].apply(list)

def get_association_rules(transactions):
    from mlxtend.preprocessing import TransactionEncoder
    from mlxtend.frequent_patterns import apriori, association_rules

    # Encode transactions
    te = TransactionEncoder()
    te_ary = te.fit(transactions).transform(transactions)
    df = pd.DataFrame(te_ary, columns=te.columns_)

    # Generate frequent itemsets
    frequent_itemsets = apriori(df, min_support=0.02, use_colnames=True)

    if frequent_itemsets.empty:
        print("No frequent itemsets found. Consider lowering min_support.")
        return pd.DataFrame()

    # Add 'num_itemsets' column for older versions of mlxtend
    if 'num_itemsets' not in frequent_itemsets.columns:
        frequent_itemsets['num_itemsets'] = frequent_itemsets['itemsets'].apply(len)

    # Generate association rules
    rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
    return rules.sort_values('confidence', ascending=False)


# Generate rules for high-value and regular customers
high_value_rules = get_association_rules(high_value_transactions)
regular_rules = get_association_rules(regular_transactions)
```

```python
# Compare unique product associations between high-value and regular customers
unique_high_value_associations = high_value_rules[
    ~high_value_rules['antecedents'].isin(regular_rules['antecedents'])
]

# Top product combinations for high-value customers during holidays
print("Unique High-Value Customer Holiday Product Associations:")
for idx, row in unique_high_value_associations.iterrows():
    antecedents = list(row['antecedents'])
    consequents = list(row['consequents'])
    print(f"If bought {antecedents}, likely to buy {consequents}")
```

```python
# Analyze high-value customer behavior by continent
def analyze_continent_behavior(continent):
    continent_high_value = high_value_holiday_purchases[
        high_value_holiday_purchases['Continent'] == continent
    ]
    return continent_high_value['Description'].value_counts().head()

# Compare holiday product preferences across continents
for continent in ['Europe', 'North America', 'Asia']:
    print(f"\n{continent} High-Value Customer Top Products:")
    print(analyze_continent_behavior(continent))
```
