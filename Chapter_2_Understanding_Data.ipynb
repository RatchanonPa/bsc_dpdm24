## JSON

- เหมิอน dictionary ใน python
- ทำตารางซ้อนตารางได้(มี list เป็นค่าข้างในได้)
- โหลดได้ 2 แบบ
1. **จากไฟล์** (`json.load()`): ใช้เมื่อไฟล์ JSON อยู่ในเครื่องหรือที่ตั้งของไฟล์
   ```python
   import json
   with open('path_to_file.json', 'r') as f:
       data = json.load(f)
   ```

2. **จากข้อความ (string)** (`json.loads()`): ใช้เมื่อข้อมูล JSON อยู่ในรูปแบบของสตริง
   ```python
   import json
   json_string = '{"key": "value"}'
   data = json.loads(json_string)
   ```



````python
# prompt: load Json file name data_station from drive
import json
json_file_path = '/content/drive/MyDrive/bsc_dpdm24_data/data_station.json'  # Replace with your actual file path

try:
    with open(json_file_path, 'r') as f:
        data = json.load(f)
        df_json = pd.DataFrame(data)  # Assuming your JSON represents a list of dictionaries
        print(df_json.head())
except FileNotFoundError:
    print(f"Error: File not found at {json_file_path}")
except json.JSONDecodeError:
    print(f"Error: Invalid JSON format in {json_file_path}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
df_json.head()
````

# Distance

````python
# prompt: build pd dataframe from [[1,2], [-1,4], [0,3], [-3,-2]] column 1 name X1 column 2 name X2

import pandas as pd

data = [[1, 2], [-1, 4], [0, 3], [-3, -2]]
df = pd.DataFrame(data, columns=['X1', 'X2'])
df
````

````python
# prompt: draw scatter plot of the df, draw bold line for x1 and x2 axes

import matplotlib.pyplot as plt

# Assuming df is already defined from the previous code

plt.figure(figsize=(8, 6))  # Adjust figure size as needed
plt.scatter(df['X1'], df['X2'], c='r')

# Customize axes with bold lines
plt.axhline(0, color='black', linewidth=2)  # X-axis line
plt.axvline(0, color='black', linewidth=2)  # Y-axis line

plt.xlabel('X1', fontweight='bold')
plt.ylabel('X2', fontweight='bold')
plt.title('Scatter Plot of X1 vs X2')
plt.grid(True)
plt.show()
````

````python
# prompt: calculate distance matrix of the df

import numpy as np
from scipy.spatial.distance import pdist, squareform

# Assuming df is your DataFrame with columns 'X1' and 'X2'
# Calculate the distance matrix
distance_matrix = squareform(pdist(df[['X1', 'X2']]))

distance_matrix
````

````python
# prompt: calculate distance matrix of the df use L1-norm

# Calculate the distance matrix using L1 norm
distance_matrix_l1 = squareform(pdist(df[['X1', 'X2']], 'cityblock'))

distance_matrix_l1
````
