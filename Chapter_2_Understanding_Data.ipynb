## JSON

- เหมิอน dictionary ใน python
- ทำตารางซ้อนตารางได้(มี list เป็นค่าข้างในได้)
- โหลดได้ 2 แบบ
1. **จากไฟล์** (`json.load()`): ใช้เมื่อไฟล์ JSON อยู่ในเครื่องหรือที่ตั้งของไฟล์
   ```python
   import json
   with open('path_to_file.json', 'r') as f:
       data = json.load(f)
   ```

2. **จากข้อความ (string)** (`json.loads()`): ใช้เมื่อข้อมูล JSON อยู่ในรูปแบบของสตริง
   ```python
   import json
   json_string = '{"key": "value"}'
   data = json.loads(json_string)
   ```



````python
# prompt: load Json file name data_station from drive
import json
json_file_path = '/content/drive/MyDrive/bsc_dpdm24_data/data_station.json'  # Replace with your actual file path

try:
    with open(json_file_path, 'r') as f:
        data = json.load(f)
        df_json = pd.DataFrame(data)  # Assuming your JSON represents a list of dictionaries
        print(df_json.head())
except FileNotFoundError:
    print(f"Error: File not found at {json_file_path}")
except json.JSONDecodeError:
    print(f"Error: Invalid JSON format in {json_file_path}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
df_json.head()
````

# Normalization (Z-score & Min-Max normalization)

## Z-score

- ปรับ data ให้อยู่ในช่วง -1,1

````python
# prompt: calculate z-score for df2

from scipy.stats import zscore

# Calculate z-scores for all columns in df2
df2_zscore = df2.apply(zscore)

df2_zscore
````

````python
# Create the scatter plot
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
#plt.plot([-3,1],[0,0], color='black', linewidth=2)
#plt.plot([0,0],[-2,4], color='black', linewidth=2)
plt.plot([-1.5,1],[0,0], color='black', linewidth=2)
plt.plot([0,0],[-1.5,1], color='black', linewidth=2)
plt.scatter(df2_zscore['X1'], df2_zscore['X2'],c='r')

min_val = min(df2_zscore['X1'].min(), df2_zscore['X2'].min())
max_val = max(df2_zscore['X1'].max(), df2_zscore['X2'].max())
plt.xlim(min_val, max_val)
plt.ylim(min_val, max_val)

# Customize the plot
plt.xlabel('X1', fontweight='bold')
plt.ylabel('X2', fontweight='bold')
plt.title('Scatter Plot of X1 vs X2')
plt.grid(True)

# Show the plot
plt.show()
````

````python
# prompt: do min-max normalization for the df2 for each column separately

# Min-Max normalization for each column of df2
df2_min_max = df2.copy()  # Create a copy to avoid modifying the original DataFrame

for column in df2.columns:
    min_val = df2[column].min()
    max_val = df2[column].max()
    df2_min_max[column] = (df2[column] - min_val) / (max_val - min_val)

df2_min_max
````

````python
# Create the scatter plot
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
plt.plot([-0.5,1.5],[0,0], color='black', linewidth=2)
plt.plot([0,0],[-0.5,1.5], color='black', linewidth=2)
plt.scatter(df2_min_max['X1'], df2_min_max['X2'],c='r')


# Customize the plot
plt.xlabel('X1', fontweight='bold')
plt.ylabel('X2', fontweight='bold')
plt.title('Scatter Plot of X1 vs X2')
plt.grid(True)

# Show the plot
plt.show()
````
